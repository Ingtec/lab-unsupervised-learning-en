# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Definir diferentes valores de k para probar
cluster_values = [2, 3, 4, 5, 6]

# Crear figura para visualizar los clusters
plt.figure(figsize=(15, 8))

for i, k in enumerate(cluster_values, 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(customers_scale)  # Asegúrate de que customers_scale está definido

    plt.subplot(2, 3, i)
    # Replace 'customers' with 'df' or 'clean_customers'
    sns.scatterplot(x=df["Grocery"], y=df["Fresh"], hue=labels, palette='viridis', alpha=0.7)
    plt.title(f'K-Means Clustering (k={k})')

plt.tight_layout()
plt.show()
```

```python
# Inicializar listas para guardar los valores de inercia y coeficiente de silueta
inertia = []
silhouette_scores = []
k_range = range(2, 10)  # Probamos valores de k desde 2 hasta 9

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(customers_scale)

    inertia.append(kmeans.inertia_)  # Suma de distancias al centroide (método del codo)
    silhouette_scores.append(silhouette_score(customers_scale, labels))  # Calidad de los clusters

# Graficar el método del codo
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')

# Graficar el coeficiente de silueta
plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o', linestyle='--', color='red')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal K')

plt.tight_layout()
plt.show()
```

- The **Silhouette Score** is highest for k = __, meaning better-defined clusters.
- Visual inspection of scatter plots suggests that __ clusters represent the data well.

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
# Definir diferentes valores de eps y min_samples para probar
eps_values = [0.2, 0.5, 1.0]
min_samples_values = [3, 5, 10]

# Crear figura para visualizar los clusters
fig, axes = plt.subplots(len(eps_values), len(min_samples_values), figsize=(15, 10))

for i, eps in enumerate(eps_values):
    for j, min_samples in enumerate(min_samples_values):
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(customers_scale)  # Asegúrate de que customers_scale está definido

        ax = axes[i, j]
        # Replace 'customers' with 'df' (original data)
        sns.scatterplot(x=df["Grocery"], y=df["Fresh"], hue=labels, palette='viridis', alpha=0.7, ax=ax)
        ax.set_title(f'DBSCAN (eps={eps}, min_samples={min_samples})')

plt.tight_layout()
plt.show()
```

```python
# Probar diferentes configuraciones y contar los clusters generados
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(customers_scale)

        unique_clusters = len(set(labels)) - (1 if -1 in labels else 0)  # Excluye outliers (-1)
        print(f"DBSCAN (eps={eps}, min_samples={min_samples}) -> Clusters: {unique_clusters}")
```

### Observations:

- When **eps is too small**, most points are considered noise (-1), meaning DBSCAN fails to cluster them.
- When **eps is too large**, DBSCAN merges most points into a single cluster, reducing effectiveness.
- **min_samples = 5 or 10** works well in most cases, balancing noise filtering and cluster formation.
- The best configuration for our dataset appears to be **eps = __ and min_samples = __**.

💡 Conclusion: Based on the analysis, we recommend using eps = __ and min_samples = __ for optimal clustering.
